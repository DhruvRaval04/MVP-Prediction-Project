```python
##using label encoding to encode two columns (Player + MVP) together - so that they both are based on the same numerical system 
## to give the model all numnerical data 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['Player', 'MVP']
df_calculationstrain[cols] = df_calculationstrain[cols].apply(LabelEncoder().fit_transform)
##df_calculationstrain['Player']=le.fit_transform(df_calculationstrain['Player'])
##df_calculationstrain['MVP']=le.fit_transform(df_calculationstrain['MVP'])


df_calculationstest[cols]= df_calculationstest[cols].apply(LabelEncoder().fit_transform)
##df_calculationstest['MVP']=le.fit_transform(df_calculationstest['MVP'])
##df_calculationstest

# df_calculationstest['Player']=le.fit_transform(df_calculationstest['Player'])
# df_calculationstest['MVP']=le.fit_transform(df_calculationstest['MVP'])
# df_calculationstest

```

```python
##cleaning the data - converting all strings to float types (numbers in the dataframe were considered strings beforehand)
##ex: Games played was '87' not 87
##df_calculationstrain['G'] = df_calculationstrain['G'].astype(float)
##df_calculationstrain = df_calculationstrain.astype({ 'MP': 'float' })
for col in df_calculationstrain.columns: 
    df_calculationstrain[col]=pd.to_numeric(df_calculationstrain[col], errors='coerce')

for col in df_calculationstest.columns: 
    df_calculationstest[col]=pd.to_numeric(df_calculationstest[col], errors='coerce')

##converting the label encoding values to float64 from int64
##df_calculationstrain = df_calculationstrain.astype(float)
df_calculationstrain['Player'] = df_calculationstrain['Player'].astype(float)
df_calculationstrain['MVP'] = df_calculationstrain['MVP'].astype(float)
df_calculationstrain = df_calculationstrain.fillna(0)

df_calculationstest['Player'] = df_calculationstest['Player'].astype(float)
df_calculationstest['MVP'] = df_calculationstest['MVP'].astype(float)

##replace all null values with 0 
##null values exist because in the early years, some stats such as steals were not recorder by the NBA 
df_calculationstest = df_calculationstest.fillna(0)
```

```python
##creating a decision tree model
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
##use the MVP column as the Y 
X_train = df_calculationstrain.drop(columns =['MVP'])
Y_train = df_calculationstrain['MVP']
X_test = df_calculationstest.drop(columns =['MVP'])
Y_test = df_calculationstest['MVP']
model = DecisionTreeClassifier()
##data = {'Name':[1,2,3],'Age':[21,22,23]}
##data2 = {'skill': [45, 46, 47]}
# dfX = pd.DataFrame(data)
# dfY = pd.DataFrame(data2)
model.fit(X_train,Y_train)
predictions = model.predict(X_test)
score = accuracy_score(Y_test, predictions)
print(score)
```
